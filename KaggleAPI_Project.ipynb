### extract data from file

```python
df = pd.read_csv('orders.csv' , na_values=['Not Available','unknown']) #Values with Not Available and unknown will be turned to null or na 
df.head()
```

```python
df['Ship Mode'].unique()
```

```python
#changing column names
# df.rename(columns={'Order Id':'order_id'}) => not a got way so many repetitions
```

```python
df.columns = df.columns.str.replace(' ','_')
df.columns = df.columns.str.lower()
df.head()
```

```python
#discount = costprice - listprice
df['discount'] = df['discount_percent']*0.01*df['list_price']

#sale price
df['sale_price'] = df['list_price'] - df['discount']

#profit
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head()
```

```python
#change order_date : object => date
df['order_date'] = pd.to_datetime(df['order_date'] , format = "%Y-%m-%d")
```

```python
#deleteing cost_price , list_price , discount_percent
df.drop(columns = ['cost_price' , 'list_price' , 'discount_percent'] , inplace = True )
```
